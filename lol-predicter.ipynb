```python
import pandas as pd
```

```python
df = pd.read_csv("cleaned_data.csv", index_col=0)
```

```python
df
```

```python
removed_columns = ['gameid', 'league', 'year', 'date', 'patch', 'teamname', 'result', 'target']
selected_columns = df.columns[~df.columns.isin(removed_columns)]
selected_columns
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[selected_columns] = scaler.fit_transform(df[selected_columns])
```

```python
df_rolling = df[list(selected_columns) + ['result', 'teamname', 'year']]
df_rolling
```

```python
def find_rolling_average(team):
    rolling = team[list(selected_columns) + ['result']].rolling(10).mean()
    return rolling

df_rolling = df_rolling.groupby(['teamname', 'year'], group_keys=False).apply(find_rolling_average)
df_rolling
```

```python
rolling_cols = [f"{col}_last10" for col in df_rolling.columns]
df_rolling.columns = rolling_cols

df = pd.concat([df, df_rolling], axis=1)
```

```python
df = df.dropna()
```

```python
df
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import RidgeClassifier

rr = RidgeClassifier(alpha=1)
split = TimeSeriesSplit(n_splits=3)

sfs = SequentialFeatureSelector(rr, n_features_to_select=10, direction='forward', cv=split)
```

```python
df
```

```python
#use if considering rolling data
selected_cols2 = df.columns[~df.columns.isin(removed_columns)]
selected_cols2
```

```python
sfs.fit(df[selected_columns], df['target'])
```

```python
predictors = list(selected_columns[sfs.get_support()])
predictors
```

```python
def backtest(data, model, predictors, start=2, step=1):
    all_predictions = []

    seasons = sorted(data['year'].unique())
    for i in range(start, len(seasons), step):
        current = seasons[i]
        train = data[data['year'] < current]
        test = data[data['year'] == current]

        model.fit(train[predictors], train['target'])
        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test['target'], preds], axis=1)

        combined.columns = ['actual', 'predictions']

        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(df, rr, predictors, start=4)
predictiontest = predictions[predictions["actual"] != 2]
predictions
```
