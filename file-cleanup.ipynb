```python
import pandas as pd
```

```python
years = [i for i in range(2016, 2024)]
```

```python
def clean_file(orig_dir, year):
    path = f"{orig_dir}/{year}_LoL_esports_match_data_from_OraclesElixir.csv"
    df = pd.read_csv(path, index_col=0)
    df = df[df['datacompleteness'].str.contains('complete')]
    df = df[df['position'].str.contains('team')]
    del df['datacompleteness']
    del df['participantid']
    del df['position']
    del df['teamkills']
    df = df.dropna(subset=['teamname', 'damagetochampions', 'dragons', 'goldat15', 'minionkills', 'visionscore', 'heralds'])
    df = df[~df['teamname'].str.contains('unknown team')]
    for s in ['quadrakills', 'pentakills', 'chemtechs', 'hextechs', 'elders', 'opp_elders', 'firstmidtower', 'firstbaron', 'inhibitors', 'opp_inhibitors']:
        df[s][pd.isnull(df[s])] = 0
    df = df.sort_values('date')
    df = df.reset_index()
    return df
```

```python
# df19 = clean_file('data', 2019)
# df20 = clean_file("data", 2020)
# df21 = clean_file("data", 2021)
# df22 = clean_file("data", 2022)
df = clean_file("data", 2023)
years.reverse()
for year in years:
    df_temp = clean_file("data", year)
    df = pd.concat([df_temp, df], ignore_index=True)
```

```python
# df = pd.concat([df19, df20, df21, df22, df23], ignore_index=True)
df = df.sort_values('date').reset_index(drop=True)
df
```

```python
nulls = pd.isnull(df)
nulls = nulls.sum()
nulls = nulls[nulls > 0]
nulls
```

```python
valid_columns = df.columns[~df.columns.isin(nulls.index)]
valid_columns
```

```python
df = df[valid_columns].copy()
df
```

```python
result = df.pop('result')
df.insert(len(df.columns), 'result', result)
```

```python
def add_target(team):
    team['target'] = team['result'].shift(-1)
    return team

df = df.groupby('teamname', group_keys=False).apply(add_target)
```

```python
df[df['teamname'] == 'T1']
```

```python
df['target'][pd.isnull(df['target'])] = 2
df['target'] = df['target'].astype(int, errors='ignore')
df
```

```python
df['side'].replace({'Blue': 1}, inplace=True)
df['side'].replace({'Red': 0}, inplace=True)
df
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import RidgeClassifier

rr = RidgeClassifier(alpha=1)
split = TimeSeriesSplit(n_splits=3)

sfs = SequentialFeatureSelector(rr, n_features_to_select=30, direction='forward', cv=split)
```

```python
removed_columns = ['gameid', 'league', 'year', 'date', 'patch', 'teamname', 'result', 'target']
selected_columns = df.columns[~df.columns.isin(removed_columns)]
selected_columns
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[selected_columns] = scaler.fit_transform(df[selected_columns])
```

```python
df
```

```python
sfs.fit(df[selected_columns], df['target'])
```

```python
predictors = list(selected_columns[sfs.get_support()])
predictors
```

```python
def backtest(data, model, predictors, start=2, step=1):
    all_predictions = []

    seasons = sorted(data['year'].unique())
    for i in range(start, len(seasons), step):
        current = seasons[i]
        train = data[data['year'] < current]
        test = data[data['year'] == current]

        model.fit(train[predictors], train['target'])
        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test['target'], preds], axis=1)

        combined.columns = ['actual', 'predictions']

        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(df, rr, predictors)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(predictions['actual'], predictions['predictions'])
```
